```python
from scipy import signal
import matplotlib.pyplot as plt
import numpy as np
import json
```

```python
path = './05.台南北華里/004x/WS_2022-10-06_11-01-12-705_Acc.json'
with open(path) as file:
    data = json.load(file)
s = np.array(data['LeftFootAcc']['x'])
print(s)
```

```python
# print(len(signal))
for i in range(1,6,1):
    b, a = signal.butter(i, 60, 'low', analog=True)
    w, h = signal.freqs(b, a)
    plt.semilogx(w, 20*np.log10(abs(h)),label = f'N = {i}')
    plt.title('Butterworth filter frequency response')
    plt.xlabel('Frequency [radians / second]')
    plt.ylabel('Amplitude [dB]')
    #plt.margins(0, 0.1)
plt.grid(which='both', axis='both')
plt.xlim(0,1000)
plt.ylim(-60,5)
# plt.axvline(100, color='green') # cutoff frequency
plt.legend()
plt.show()
```
